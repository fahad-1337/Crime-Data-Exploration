# **EDA**

```python
#What are some common crimes in terms of offense groups?

df.OFFENSE_CODE_GROUP.value_counts()
```

```python
offense_group_top = df.OFFENSE_CODE_GROUP.value_counts()[0:10]

df.shape[0]

OF_Group_percentage = offense_group_top / df.shape[0]*100

OF_Group_percentage.plot(kind='bar')
plt.title("Top 10 Offense Groups(% of all crimes)")
```

```python
# What are the most common offense descriptions?

df.OFFENSE_DESCRIPTION.value_counts()[0:5]

(df.OFFENSE_DESCRIPTION.value_counts()[0:5]/df.shape[0]*100).plot(kind='bar')
plt.title("Top 5 Offense Descriptions(% of all crimes)")


```

```python
#Number of crimes in each year?
df.groupby('YEAR').count()['INCIDENT_NUMBER'].plot(kind='bar')
```

```python
#Are they more crimes comitted on specific days?

df.groupby('DAY_OF_WEEK').count()['INCIDENT_NUMBER'].plot(kind='bar')
```

```python
df.groupby('HOUR').count()['INCIDENT_NUMBER'].plot(kind='bar')
plt.title("Number of Crimes per Hour")
```

```python
#On What days and which hours are most crimes committed?
```

```python
df.groupby(['HOUR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()
```

```python
hour_and_week = df.groupby(['HOUR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()
```

```python
hour_and_week.columns = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
hour_and_week.columns
```

```python
sns.heatmap(hour_and_week,cmap='YlGnBu')
```

```python
#How many crimes occured in each month?
```

```python
most_crimes = df.groupby(['MONTH','YEAR']).count()['INCIDENT_NUMBER'].unstack()
most_crimes
```

```python
#In which district were the most crimes committed on a daily basis?

distrcitwise_crimes = df.groupby(['DISTRICT','YEAR']).count()['INCIDENT_NUMBER'].unstack()

sns.heatmap(distrcitwise_crimes,cmap='YlGnBu')
```
